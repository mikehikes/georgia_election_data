## Finding Repeat Voters

With the voting data now organized by voter and by voting record, useful statistics may now be calculated. We will start with identifying the characteristics of "repeat" voters: voters who vote again, after voting in a previous election. 

Results of this statistic can be helpful for purposes of identifying voter engagement and therefore modeling voter behavior in future elections.


### Expected Behavior

The following stylized examples describe the expected intuition of "repeat" voting behavior:

#### Voter A
> Voter A first voted in the 1996 General Election; this voter then voted again in 1998, and 2000, but died in 2001, and therefore could not vote in subsequent elections.

#### Voter B
> Voter B first voted in the 2000 Primary, and continued voted until the 2016 Election; however, Voter B missed the 2002 and 2004 elections, since they forgot to request an absentee ballot.

#### Voter C
> Voter C has voted sporadically since 2010, based on their personal interest in the candidates and/or election.

#### Voter D
> Voter D has voted sporadically since 2002 until 2012, at which time they moved outside the State.

### Hypothesis

For purposes of this analysis, We can assume that for each cohort of voters who vote in an earlier election, a smaller percentage of voters will vote again in a future election due to three factors: 

- First, those are able/eligible but choose not to vote in a subsequent election
- Second, those voters who, through death, disability, conviction of a felony, other life-altering experiences, are unable or ineligible to to vote.
- Third, those voters who eligible but not able to vote, due to bureaucratic, logistical, or other reasons of which the voter cannot directly impact. 

This analysis will not directly evaluate this hypothesis, but rather will calculate aggregate statistics that may provide insights to exploring this question.

Specifically, this notebook will calculate for various combinations of voter and election cohorts, the aggregate participation levels of those cohorts.

## Analysis of Repeat Voters

Finally we have the data prepared to conduct some descriptive analysis! First, let's calculate for primary and general elections, how many people vote in each years'velections. Now that the data has been aggregated we can also calculate what percentage of voters vote one year or type of election and then another.

```python
"""
This function iterates through each voter in our voter database, incrementing a counter
each time a voter votes in one or both elections as defined in the tuple_list.
"""

def election_participation_counter(in_dict, tuple_list=None, source_cat=None, dest_cat=None):
    
    tuple_dict = {}
    for t in tuple_list:
        tuple_dict.update({t:{'source_count':0,
                             'dest_count':0,
                             'both_count':0}})
    
    for k in tqdm(in_dict.keys(),position=0):
        v = in_dict[k]
        
        for tuple_key, tuple_item in tuple_dict.items():
            source_year, dest_year = tuple_key
            
            if source_year in v[source_cat]:
                tuple_dict[tuple_key]['source_count'] += 1
            if dest_year in v[dest_cat]:
                tuple_dict[tuple_key]['dest_count'] += 1
            if source_year in v[source_cat] and dest_year in v[dest_cat]:
                tuple_dict[tuple_key]['both_count'] += 1
            
    return tuple_dict
```

We will now define the election pairs that will be utilized by the `election_participation_counter`.

As an example, if the election pair is `(2000, 2004)`, this will result in the `election_participation_counter` determining first, how many voters have voted in the _2000_ election; second, how many voters have voted in the _2004_ election, and how many voters voted in both the _2000_ and _2004_ election.

```python
# List of years' elections we are evaluating
year_list = ['1996','1998','2000','2002','2004','2006','2008','2010','2012','2014','2016','2018']
year_list_combinations = list(product(year_list, year_list))

# For now, the dest year should always be greater or equal to the source year
# It is 
year_list_combinations = list(filter(lambda x: x[1] >=x[0], year_list_combinations))

print(f"{len(year_list_combinations)} year combinations generated. \n\
The first combination is {year_list_combinations[0]} and the last one is {year_list_combinations[-1]}") 
 
```

## Initiate the calculations

We will now run the calculations; each calculation involves a full iteration through the `voter_dict` shelf object, and should take about 5-6 minutes (as measured on an Intel Core i7).  

```python
# We will calculate general to general, primary (partisan and non-partisan) to general, 
# primary_democrat to general, and primary_republican to general. 

general_to_general = election_participation_counter(voter_dict, year_list_combinations, 'g','g')
print("Finished general_to_general")
primary_to_general = election_participation_counter(voter_dict, year_list_combinations, 'p','g')
print("Finished primary_to_general")
primary_to_primary = election_participation_counter(voter_dict, year_list_combinations, 'p','p')
print("Finished primary_to_primary")
primary_democrat_to_general = election_participation_counter(voter_dict, year_list_combinations, 'd_p','g')
print("Finished primary_democrat_to_general")
primary_republican_to_general = election_participation_counter(voter_dict, year_list_combinations, 'r_p','g')
print("Finished primary_republican_to_general")
print("Done!")
```

For ease of access, we will create a new dictionary with the results of the above calculation. We will also save the results as a pickle object for future access so that this notebook may be re-used without the need to spend time recalculating the above statistics.

```python
ret_dict = {'general_to_general':general_to_general,
            'primary_to_general':primary_to_general,
            'primary_to_primary':primary_to_primary,
            'primary_democrat_to_general':primary_democrat_to_general,
            'primary_republican_to_general':primary_republican_to_general}
pickle.dump(ret_dict, open('election_participation_counter.pickle','wb'))
```

```python
# If the above statistics were successful, 
# then the following command will reload the pickle object that was just created.

ret_dict = pickle.load(open('election_participation_counter.pickle','rb'))
```

Looking at the results of one of these participation counters...

```python
example_source = ret_dict['general_to_general'][('2000','2004')]['source_count']
example_dest = ret_dict['general_to_general'][('2000','2004')]['dest_count']
example_both = ret_dict['general_to_general'][('2000','2004')]['both_count']
print(f"Example slice of results: {ret_dict['general_to_general'][('2000','2004')]}")
```

...we find three numbers: the `source_count` indicates the number of voters (2,637,040) who voted in the 2000 general election; the `dest_count` indicates the number of voters (3,290,827) who voted in the 2004 general election, and the `both_count` indicates the number of voters who voted in both the 2000 and 2004 general presidential elections.

```python
print(f"{np.round(example_both/example_source,2)*100}% of voters in the 2000 Presidential Election voted again in the 2004 Presidential Election")
print(f"{np.round(example_both/example_dest,2)*100}% of voters in the 2004 Presidential Election voted earlier in the 2000 Presidential Election")
```

For now, we are only going to map the `source_count` to the `dest_count`, and not vice-versa

```python
def calculate_pct_participation(in_dict=None, in_dict_data_key=None, dest_year_mask=[]):
    
    df = pd.DataFrame(ret_dict[in_dict_data_key]).transpose()
    df['both_div_source'] = df.apply(lambda x: x.both_count / x.source_count, axis=1)
    df = df[['both_div_source']]
    df = df.reset_index()
    df.columns = ['source_year','dest_year','pct_participated']
    df = df.loc[df.apply(lambda x: x.dest_year in dest_year_mask and x.pct_participated < 1 ,axis=1)]

    return df
```

```python
gen_to_gen = calculate_pct_participation(in_dict = ret_dict, in_dict_data_key = 'general_to_general',
                                        dest_year_mask=['1996','2000','2004','2008','2012','2016'])

print("Sample of data collected:")
gen_to_gen.head()
```

### Data Visualization Strategies

Pandas/matplotlib plots have the advantage of being able to be saved within a notebook; however, they have the disadvantage of being static - preventing exploratory data analysis of the results.

Plotly charts are interactive; however, they cannot presently be saved as output within a Jupyter notebook.

In order to verify the proper operation of code, the matplotlib-based graphics are included; however, the analyst may generate Plotly charts for purposes of interactive exploration of the results.

Before plotting the results, the following is a utility function for our Plotly charts.

```python
pres_years = ['1996','2000','2004','2008','2012','2016']
non_pres_years = ['1998','2002','2006','2010','2014','2018']

def plot_lines(x,in_fig):
    
    # generate hover text
    
    pct_participated_list = x.pct_participated.tolist()
    source_year_list = x.source_year.tolist()
    dest_year_list = x.dest_year.tolist()
    input_tuples = list(zip(pct_participated_list, source_year_list, dest_year_list))
    text = [f"{np.round(a*100,2)} % of voters who voted in {b} voted again in {c}" for a,b,c in input_tuples]
    
    if source_year_list[0] in non_pres_years:
        in_fig.add_trace(go.Scatter(x=x.dest_year, y=x.pct_participated,
                        mode='lines+markers',
                        hovertext = text,
                        hoverinfo= "text",
                        line = dict(dash="dash"),
                        name=x.source_year.iloc[0]))
        
    else:
        
        in_fig.add_trace(go.Scatter(x=x.dest_year, y=x.pct_participated,
                        mode='lines+markers',
                        hovertext = text,
                        hoverinfo= "text",
                        name=x.source_year.iloc[0]))        
        
    pass
    
```

#### This set of charts indicate, for each general election - both in Presidential- and non-Presidential years, the percentage of voters who voted again in a Presidential election.

```python
# Matplotlib version
gen_to_gen_plt = deepcopy(gen_to_gen)
fig, ax = plt.subplots()
gen_to_gen_plt.source_year = gen_to_gen_plt.source_year.apply(lambda x: int(x))
gen_to_gen_plt.dest_year = gen_to_gen_plt.dest_year.apply(lambda x: int(x))
gen_to_gen_plt.sort_values(['source_year','dest_year'],ascending=True).groupby('source_year').plot(fontsize=14, figsize=(15,7),marker='o',x='dest_year' ,y='pct_participated', ax=ax)
ax.legend(gen_to_gen_plt.source_year.drop_duplicates())
ax.set_xlabel('Year that the voter voted again', fontsize=16)
ax.set_ylabel('Percentage of voters who voted again,\n by first voting year', fontsize=16)
ax.set_xlim(1999, 2017)
```

```python
#Plotly version
g2g_fig = go.Figure()
_ = gen_to_gen.groupby('source_year').apply(lambda x: plot_lines(x=x, in_fig=g2g_fig))

g2g_fig.update_layout(title='For each general election year, % of voters voting again in a presidential general election <br>Dashed lines are non-presidential general elections',
                   xaxis_title='Year that the voter voted again',
                   yaxis_title='Percentage of voters that voted again')

g2g_fig.show()
```
